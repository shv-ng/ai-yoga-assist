Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Bidirectional, LSTM, Dense, Dropout
from tensorflow.keras.utils import to_categorical

print("TensorFlow version:", tf.__version__)
print("All libraries loaded ✅")
```

Load the CSV and group into sequences

```python
df = pd.read_csv("poses.csv")
feature_cols = [c for c in df.columns if c not in ["frame_id", "sequence_id", "label"]]

print("Dataset shape:", df.shape)
print("\nSamples per pose:")
print(df["label"].value_counts())
print("\nUnique sequences:", df["sequence_id"].nunique())
print("\nAverage frames per sequence:")
print(df.groupby("sequence_id").size().describe())
```

Group frames into sequences of 30 frames

```python
SEQUENCE_LENGTH = 30

sequences = []
sequence_labels = []

for label in df["label"].unique():
    pose_frames = df[df["label"] == label][feature_cols].values  # all frames for this pose

    # slide through frames, take chunks of 30
    for i in range(0, len(pose_frames) - SEQUENCE_LENGTH + 1, SEQUENCE_LENGTH):
        seq = pose_frames[i : i + SEQUENCE_LENGTH]  # shape (30, 99)
        sequences.append(seq)
        sequence_labels.append(label)

X_seq = np.array(sequences)   # shape (N, 30, 99)
y_seq = np.array(sequence_labels)

print("Sequences shape:", X_seq.shape)
print("Labels shape   :", y_seq.shape)
print("\nSequences per pose:")
for label in np.unique(y_seq):
    print(f"  {label}: {np.sum(y_seq == label)}")
```

Generate "Incorrect" versions of each sequence

```python
def make_incorrect(sequence, noise=0.15):
    """Perturb a correct sequence to simulate wrong form."""
    corrupted = sequence.copy()

    # Randomly shift some landmarks
    num_landmarks = corrupted.shape[1] // 3
    landmarks_to_perturb = np.random.choice(num_landmarks, size=8, replace=False)

    for lm in landmarks_to_perturb:
        idx = lm * 3
        corrupted[:, idx]   += np.random.uniform(-noise, noise)  # x
        corrupted[:, idx+1] += np.random.uniform(-noise, noise)  # y
        corrupted[:, idx+2] += np.random.uniform(-noise, noise)  # z

    return np.clip(corrupted, 0, 1)

# Create correct + incorrect dataset
X_correct = X_seq.copy()
y_correct = np.ones(len(X_seq))   # 1 = correct form

X_incorrect = np.array([make_incorrect(seq) for seq in X_seq])
y_incorrect = np.zeros(len(X_seq))  # 0 = incorrect form

# Combine
X_all = np.concatenate([X_correct, X_incorrect], axis=0)  # (24, 30, 99)
y_all = np.concatenate([y_correct, y_incorrect], axis=0)  # (24,)

print("Total sequences  :", len(X_all))
print("Correct samples  :", int(y_correct.sum()))
print("Incorrect samples:", int(np.sum(y_incorrect == 0)))
print("X shape          :", X_all.shape)
```

Visualise the difference between correct and incorrect

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 4))

# pick one sequence to compare
correct_seq   = X_correct[0]    # shape (30, 99)
incorrect_seq = X_incorrect[0]  # shape (30, 99)

# plot landmark 0 (nose x position) across 30 frames
axes[0].plot(correct_seq[:, 0],   label="Correct",   color="steelblue")
axes[0].plot(incorrect_seq[:, 0], label="Incorrect", color="coral", linestyle="--")
axes[0].set_title("Landmark 0 (Nose X) — Correct vs Incorrect")
axes[0].set_xlabel("Frame")
axes[0].set_ylabel("Position value")
axes[0].legend()
axes[0].grid(True)

# plot the difference between them
diff = np.abs(correct_seq - incorrect_seq)
axes[1].imshow(diff.T, aspect="auto", cmap="hot")
axes[1].set_title("Heatmap of Differences (all 99 landmarks × 30 frames)")
axes[1].set_xlabel("Frame")
axes[1].set_ylabel("Landmark value")
plt.colorbar(axes[1].images[0], ax=axes[1])

plt.tight_layout()
plt.show()
```

Split data & Build the BiLSTM

```python
# Split
X_train, X_test, y_train, y_test = train_test_split(
    X_all, y_all, test_size=0.2, random_state=42
)

print("Train sequences:", len(X_train))
print("Test sequences :", len(X_test))

# Build model
model_bilstm = Sequential([
    Bidirectional(LSTM(64, return_sequences=True), input_shape=(30, 99)),
    Dropout(0.3),
    Bidirectional(LSTM(32)),
    Dropout(0.3),
    Dense(16, activation="relu"),
    Dense(1, activation="sigmoid")   # binary output: correct or incorrect
], name="PoseCorrector")

model_bilstm.compile(
    optimizer="adam",
    loss="binary_crossentropy",
    metrics=["accuracy"]
)

model_bilstm.summary()
```

Train the BiLSTM

```python
history = model_bilstm.fit(
    X_train, y_train,
    validation_data=(X_test, y_test),
    epochs=50,
    batch_size=8,
    verbose=1
)
```

Visualise Training + Evaluate

```python
# Training curves
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].plot(history.history["accuracy"], label="Train Accuracy", color="steelblue")
axes[0].plot(history.history["val_accuracy"], label="Val Accuracy", color="coral")
axes[0].set_title("BiLSTM Accuracy over Epochs")
axes[0].set_xlabel("Epoch")
axes[0].set_ylabel("Accuracy")
axes[0].legend()
axes[0].grid(True)

axes[1].plot(history.history["loss"], label="Train Loss", color="steelblue")
axes[1].plot(history.history["val_loss"], label="Val Loss", color="coral")
axes[1].set_title("BiLSTM Loss over Epochs")
axes[1].set_xlabel("Epoch")
axes[1].set_ylabel("Loss")
axes[1].legend()
axes[1].grid(True)

plt.tight_layout()
plt.show()

# Evaluation
print("=" * 40)
y_pred = (model_bilstm.predict(X_test, verbose=0) > 0.5).astype(int)
print(classification_report(y_test.astype(int), y_pred, target_names=["Incorrect", "Correct"]))
```

Save the BiLSTM

```python
import os
os.makedirs("models", exist_ok=True)

model_bilstm.save("models/pose_corrector.keras")
print("BiLSTM model saved → models/pose_corrector.keras")
```
